### Mean Encoding
- Ya que en este dataset solo utilizamos variables categoricas, aplicamos un método encoding como transformacion de datos para hacer que los valores correspondientes a estas variables sean mas comprensibles para el modelo y ayudarlo a dar una mejor predicción.

```python
from sklearn.model_selection import KFold

folds = KFold(5, shuffle = False)

for train_index, val_index in folds.split(df):

    X_train, X_val = df.iloc[train_index], df.iloc[val_index]

    df.loc[df.index[val_index], 'Gender_target_enc'] = X_val['Gender'].map(\
        X_train.groupby('Gender').Target.mean()).replace(0, 0.408)
    df.loc[df.index[val_index], 'Age_target_enc'] = X_val['Age'].map(\
        X_train.groupby('Age').Target.mean()).replace(0, 0.408)
    df.loc[df.index[val_index], 'City_Cat_target_enc'] = X_val['City_Category'].map(\
        X_train.groupby('City_Category').Target.mean()).replace(0, 0.408)
    df.loc[df.index[val_index], 'Stay_In_City_target_enc'] = X_val['Stay_In_Current_City_Years'].map(\
        X_train.groupby('Stay_In_Current_City_Years').Target.mean()).replace(0, 0.408)
    df.loc[df.index[val_index], 'Product_Category_1_enc'] = X_val['Product_Category_1'].map(\
        X_train.groupby('Product_Category_1').Target.mean()).replace(0, 0.408)
    df.loc[df.index[val_index], 'Product_Category_2_enc'] = X_val['Product_Category_2'].map(\
        X_train.groupby('Product_Category_2').Target.mean()).replace(0, 0.408)
    df.loc[df.index[val_index], 'Product_Category_3_enc'] = X_val['Product_Category_3'].map(\
        X_train.groupby('Product_Category_3').Target.mean()).replace(0, 0.408)
```

```python
# Ubicamos la variable objetivo y la ponemos en una variable nueva

Target = df.Target

df.drop(['Purchase', 'Marital_Status', 'Target', 'Age'], axis = 1, inplace = True)

# Realizamos un One hot encoding para poder correr los modelos
c_cols = ['Gender', 'Occupation', 'City_Category',
   'Stay_In_Current_City_Years', 'Product_Category_1',
   'Product_Category_2', 'Product_Category_3']

dummies = pd.get_dummies(df[c_cols])
d_cols = dummies.columns
df.drop(c_cols, axis = 1, inplace = True)
df[d_cols] = dummies
```

### Funciones

```python
def feature_plot(importances, X_train, y_train):

    # Mostrar las 5 variables mas importantes
    indices = np.argsort(importances)[::-1]
    columns = X_train.columns.values[indices[:7]]
    values = importances[indices][:7]


    # Crear la gráfica
    fig = plt.figure(figsize = (9,7))
    plt.title("Pesos normalizados para las 5 variables mas predictivas", fontsize = 16)
    plt.bar(np.arange(7), values, width = 0.6, align = 'center', color = '#00A000', \
        label = 'Peso de la variable')
    plt.bar(np.arange(7) -0.3, np.cumsum(values), width = 0.2, align = 'center', color = '#00A0A0', \
        label = 'Peso cumulativo para la variable')
    plt.xticks(np.arange(5), columns)
    plt.xlim((-0.5, 4.5))
    plt.xlabel('Peso', fontsize = 12)
    plt.xlabel('Variable', fontsize = 12)

    plt.legend(loc = 'upper center')
    plt.tight_layout()
    plt.show()
```

```python
def train_predict(learner, sample_size, X_train, y_train, X_test, y_test):
    '''
    inputs:
    - learner: the learning algorithm to be trained and predicted on
    - sample_size: the size of samples (number) to be drawn from training set
    - X_train: features training set
    - y_train: income training set
    - X_test: features testing set
    - y_test: income testing set
    '''

    results = {}

    start_train = time()
    model = learner
    learner.fit(X_train[:sample_size], y_train[:sample_size])
    end_train = time()

    # Calcular el tiempo de entrenamiento
    results['train_time'] = end_train - start_train

    start_prediction = time()
    predictions_test = model.predict(X_test)
    predictions_train = model.predict(X_train[:300])
    end_prediction = time()

    # Calcular el tiempo de predicción
    results['pred_time'] = end_prediction - start_prediction

    # Calcular precisión de los 300 primeros datos
    results['acc_train'] = accuracy_score(predictions_train, y_train[:300])

    #Calcular precisión en el set de prueba 
    results['acc_test'] = accuracy_score(predictions_test, y_test)

    # Calcular F-score en los primeros 300 datos
    results['f_train'] = fbeta_score(y_train[:300], predictions_train,  beta = 0.5)

    # Calcular F-score en el set de prueba
    results['f_test'] = fbeta_score(y_test, predictions_test, beta = 0.5)

    print("{} trained on {} samples.".format(learner.__class__.__name__, sample_size))

    return results
```

```python
def evaluate(results, accuracy, f1):
    """
    Visualization code to display results of various learners.

    inputs:
      - learners: a list of supervised learners
      - stats: a list of dictionaries of the statistic results from 'train_predict()'
      - accuracy: The score for the naive predictor
      - f1: The score for the naive predictor
    """

    fig, ax = plt.subplots(2, 3, figsize = (11,7))

    bar_width = 0.3
    colors = ['#A00000','#00A0A0','#00A000']

    for k, learner in enumerate(results.keys()):
        for j, metric in enumerate(['train_time', 'acc_train', 'f_train', 'pred_time', 'acc_test', 'f_test']):
            for i in np.arange(3):

                # Codigo de gráfica creativo
                ax[j//3, j%3].bar(i+k*bar_width, results[learner][i][metric], width = bar_width, color = colors[k])
                ax[j//3, j%3].set_xticks([0.45, 1.45, 2.45])
                ax[j//3, j%3].set_xticklabels(["1%", "10%", "100%"])
                ax[j//3, j%3].set_xlabel("Training Set Size")
                ax[j//3, j%3].set_xlim((-0.1, 3.0))

    # Añadir valores de y únicos
    ax[0, 0].set_ylabel("Tiempo (en segundos)")
    ax[0, 1].set_ylabel("Precisión")
    ax[0, 2].set_ylabel("F-score")
    ax[1, 0].set_ylabel("Tiempo (en segundos)")
    ax[1, 1].set_ylabel("Precisión")
    ax[1, 2].set_ylabel("F-score")


    # Añadir títulos
    ax[0, 0].set_title("Entrenamiento")
    ax[0, 1].set_title("Precision en el set de entrenamiento")
    ax[0, 2].set_title("F-score on set de entrenamiento")
    ax[1, 0].set_title("Prediccion del modelo")
    ax[1, 1].set_title("Precision en el set de entrenamiento")
    ax[1, 2].set_title("F-score en el set de prueba")


    # Añadir lineas horizontales para los predictores ingenuos
    ax[0, 1].axhline(y = accuracy, xmin = -0.1, xmax = 3.0, linewidth = 1, color = 'k', linestyle = 'dashed')
    ax[1, 1].axhline(y = accuracy, xmin = -0.1, xmax = 3.0, linewidth = 1, color = 'k', linestyle = 'dashed')
    # ax[0, 2].axhline(y = f1, xmin = -0.1, xmax = 3.0, linewidth = 1, color = 'k', linestyle = 'dashed')
    # ax[1, 2].axhline(y = f1, xmin = -0.1, xmax = 3.0, linewidth = 1, color = 'k', linestyle = 'dashed')

    # Ajustar los ejes y para los paneles
    ax[0, 1].set_ylim((0, 1))
    ax[0, 2].set_ylim((0, 1))
    ax[1, 1].set_ylim((0, 1))
    ax[1, 2].set_ylim((0, 1))

    # crear 'parches' para la leyenda
    patches = []
    for i, learner in enumerate(results.keys()):
        patches.append(mpatches.Patch(color = colors[i], label = learner))
    plt.legend(handles = patches, bbox_to_anchor = (-.60, 1.33), \
                loc = 'center', borderaxespad = 0., ncol = 3, fontsize = 'x-large')

    # Estetica
    plt.suptitle("Metricas para evaluar tres diferentes modelos de clasificación", fontsize = 16, y = 1.10)
    plt.tight_layout(pad = 1, w_pad = 2, h_pad = 2)
    plt.show()
```

```python
def display_roc():

    clfs = {'clf_A': LogisticRegression(random_state = 42),
    'clf_B': DecisionTreeClassifier(random_state = 42),
    'clf_C': RandomForestClassifier(random_state = 42)}

    # Primer clf
    plt.subplot(2,2,1)
    clfs.get('clf_A').fit(X_train, y_train)
    y_pred_prob = clfs.get('clf_A').predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr)
    plt.xlabel('Falsos positivos')
    plt.ylabel('Verdaderos positivos')
    plt.title(f'ROC Curve {clfs.get("clf_A").__class__.__name__}: {roc_auc_score(y_test, y_pred_prob):.2f}')
    # Segundo clf
    plt.subplot(2,2,2)
    clfs.get('clf_B').fit(X_train, y_train)
    y_pred_prob = clfs.get('clf_B').predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr)
    plt.xlabel('Falsos positivos')
    plt.ylabel('Verdaderos positivos')
    plt.title(f'ROC Curve {clfs.get("clf_B").__class__.__name__}: {roc_auc_score(y_test, y_pred_prob):.2f}')
    # Tercer clf
    plt.subplot(2,2,3)
    clfs.get('clf_C').fit(X_train, y_train)
    y_pred_prob = clfs.get("clf_C").predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr)
    plt.xlabel('Falsos positivos')
    plt.ylabel('Verdaderos positivos')
    plt.title(f'ROC Curves {clfs.get("clf_C").__class__.__name__}: {roc_auc_score(y_test, y_pred_prob):.2f}')
    plt.tight_layout(pad = 1,w_pad = 1.5, h_pad = 0.8)
    plt.show()
```

### Predicciones utilizando la densidad original (Predictor ingenuo)

```python
#Calculamos el predictor ingenuo para que nos de la densidad original

TP = np.sum(Target)
print('true positives: {}'.format(TP))
FP = Target.count() - TP
print('false negatives: {}'.format(FP))
precision = TP/(TP+FP)
print('precision: {}'.format(precision))
recall = TP/TP
accuracy = precision
fscore = (1+0.5**2)*(precision * recall)/((0.5**2)*precision + recall)

print("Predictor ingenuo: [Densidad original: {}, F-score: {}]".format(accuracy, fscore))
```

### Elección del modelo
   - Elegimos tres modelos de los que vimos en clase y los probamos para poder elegir uno con el cual trabajar

```python
# Dividimos el dataset en train y test

X_train, X_test, y_train, y_test = train_test_split(df, Target, 
                                test_size = 0.3,
                                random_state = 42)
```

```python
# Extracción de variables importantes

model = RandomForestClassifier()
model.fit(X_train, y_train)
importances = model.feature_importances_
plt.show(feature_plot(importances, X_train, y_train))
```

```python
clf_A = LogisticRegression(random_state = 42)
clf_B = DecisionTreeClassifier(random_state = 42)
clf_C = RandomForestClassifier(random_state = 42)


##Calcular el numero de sets para 1%, 10%, y 100% del set de entrenamiento
samples_100 = len(X_train)
# print('samples_100: {}'.format(samples_100))
samples_10 = round(samples_100*0.10)
# print('samples_10: {}'.format(samples_10))
samples_1 = round(samples_100*0.01)
# print('samples_1: {}'.format(samples_1))

# Recolectar resultados de los modelos
results = {}
for clf in [clf_A, clf_B, clf_C ]:
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    for i, samples in enumerate([samples_1, samples_10, samples_100]):
        results[clf_name][i] = \
        train_predict(clf, samples, X_train, y_train, X_test, y_test)
```

```python
# Metricas para evaluar tres diferentes modelos de clasificación
plt.show(evaluate(results, accuracy, fscore))
```

### Ajuste de hiperparametros
- Tras observar lo anterior optamos por utilizar un Random Forest para utilizarlo como modelo final y utilizar un ajuste de hiperparametros en el.
- Lo que queremos es poder predecir por medio de la conducta de compra y otras variables si un consumidor es casado o soltero 

```python
param_grid = {
    "n_estimators": np.random.randint(15,60,10),
    "max_depth": [30, None],
    "max_features": ['sqrt', 'log2', None],
    "min_samples_leaf": np.random.randint(1,40,10),
    "criterion": ['gini', 'entropy']
}

clf = RandomForestClassifier(random_state = 42, n_jobs = -1)

clf.fit(X_train, y_train)

scorer = make_scorer(fbeta_score, beta=0.5)

clf_cv = RandomizedSearchCV(clf, param_grid, cv=4, scoring = scorer)

labels = y_train.values

time1 = time()

grid_fit = clf_cv.fit(X_train, labels)

best_clf = grid_fit.best_estimator_

predictions = clf.predict(X_test)
print(predictions.shape)
print(y_test.shape)

best_predictions = best_clf.predict(X_test)

print("Modelo sin optimizar\n------")
print("Precisión en el set de prueba: {}".format(accuracy_score(y_test, predictions)))
print("F-score en el set de prueba: {}".format(fbeta_score(y_test, predictions, beta = 0.5)))
print("\nModelo Optimizado\n------")
print("Precisión final en el set de prueba: {}".format(accuracy_score(y_test, best_predictions)))
print("F-score final en el set de prueba: {}".format(fbeta_score(y_test, best_predictions, beta = 0.5)))
print(f'Tiempo de entrenamiento: {time() - time1}')
```
